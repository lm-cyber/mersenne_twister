```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [16,10]
plt.rcParams.update({'font.size':18})
f = open('data.txt', 'r')
step =0.0001
kol = np.arange(start=0,stop=1,step=step)
mass_of_date = [float(el) for el in f.read().split()]
plt.plot(kol,mass_of_date,color='green',linewidth=1.5,label='data')
plt.legend()
```

```python
n    = len(kol)
yf   = rfft(mass_of_date)
xf   = rfftfreq(n,step)
plt.plot(xf,np.abs(yf))
```
